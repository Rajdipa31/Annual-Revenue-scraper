```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
import time
import pandas as pd
from bs4 import BeautifulSoup 

# Initialize the Selenium WebDriver
def init_driver():
    options = Options()
    options.headless = False  # Set to True for headless mode (no UI)

    # Use webdriver-manager to automatically download and set up chromedriver
    service = Service(ChromeDriverManager().install())  # Use Service to set the chromedriver path
    driver = webdriver.Chrome(service=service, options=options)  # Pass service and options
    return driver

# Search for company turnover on Google and scrape data
def get_turnover_from_google(company_name):
    driver = init_driver()
    search_query = f"{company_name} last year turnover"
    
    try:
        # Go to Google and search for the company turnover
        driver.get("https://www.google.com")
        search_box = driver.find_element(By.NAME, "q")  # Find the Google search box
        search_box.send_keys(search_query)
        search_box.send_keys(Keys.RETURN)  # Press Enter

        # Wait for the page to load
        time.sleep(3)

        # Get the page source after the search
        page_source = driver.page_source
        soup = BeautifulSoup(page_source, 'html.parser')

        # Find relevant information (e.g., from a snippet or table)
        turnover = None
        # Looking for a specific HTML element that might contain the turnover value
        for elem in soup.find_all('span'):
            if 'revenue' in elem.get_text().lower() or 'turnover' in elem.get_text().lower():
                turnover = elem.get_text()

        if turnover:
            return turnover.strip()
        else:
            return "Turnover not found"

    except Exception as e:
        print(f"Error while searching for turnover of {company_name}: {e}")
        return "Error fetching turnover"

    finally:
        driver.quit()  # Close the browser after scraping

# Create a new DataFrame to store the results
def fetch_turnover_for_companies(merged_data):
    company_turnover_data = []

    for company in merged_data['Company Name']:
        print(f"Fetching turnover for: {company}")
        turnover = get_turnover_from_google(company)
        company_turnover_data.append((company, turnover))

    turnover_df = pd.DataFrame(company_turnover_data, columns=['Company Name', 'Turnover'])
    return turnover_df

# Example usage
# merged_data = pd.read_csv('your_merged_data.csv')  # Load your merged_data DataFrame

turnover_df = fetch_turnover_for_companies(merged_data)

# Display the DataFrame with company names and turnovers
print(turnover_df)
```
